```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import LabelBinarizer
from sklearn.naive_bayes import MultinomialNB

```

```python
df = pd.read_csv("treated_df.csv", sep=';', index_col=0)
df = df.dropna()
df.head()
```

```python
### REGROUPER LES GENRE EN GRAND GENRE ###
# On part sur les tags associés

distincts = list(set(df["genre_name"].to_list()))
print(distincts)


```

```python
labels = df["genre_name"].to_list()
distincts = list(set(labels))

for i in range(len(distincts)):
    colonne = []
    for j in range(len(labels)):
        if labels[i] == distincts :
            colonne.append(1)
        else :
            colonne.append(0)
    df[distincts[i]] = colonne
                   
df
```

```python
#### TRAITEMENT DES DONNEES ####
text_data = df["lyrics"]
labels = df["genre_name"]

X_train, X_test, y_train, y_test = train_test_split(text_data, labels, test_size=0.2, random_state=42)

# Vectorisation des données textuelles en utilisant CountVectorizer
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Entraînement du classificateur Naive Bayes multinomial
classifier = MultinomialNB()
classifier.fit(X_train_vectorized, y_train)

# Prédictions sur les données de test
predictions = classifier.predict(X_test_vectorized)

# Évaluation du modèle
accuracy = accuracy_score(y_test, predictions)
print("Précision du modèle :", accuracy)
```
