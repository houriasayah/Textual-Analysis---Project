```python
def detect_language(text):
    try:
        return detect(text)
    except:
        return 'unknown'

# Appliquez la fonction à chaque ligne de la colonne 'lyrics'
df['language'] = df['lyrics'].apply(lambda x: detect_language(x))

print(df)
```
