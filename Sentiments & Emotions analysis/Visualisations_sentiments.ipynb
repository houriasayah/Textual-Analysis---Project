```python
import pandas as pd

data = pd.read_csv("sentiments_emotions_df.csv")
```

```python
print(data)
```

```python
import pandas as pd

# Supposez que votre DataFrame s'appelle data

# Trouver le score majoritaire pour chaque ligne
data['sentiment'] = data[['positive score', 'neutral score', 'negative score']].idxmax(axis=1)

# Mapper les indices des colonnes à leur étiquette respective
sentiment_map = {'positive score': 'positive', 'neutral score': 'neutral', 'negative score': 'negative'}
data['sentiment'] = data['sentiment'].map(sentiment_map)

print(data)
```

```python
data.to_csv("sentiments_emotions_df.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Histogramme pour le sentiment dominant
plt.figure(figsize=(8, 6))
data['sentiment'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution des Sentiments Dominants')
plt.xlabel('Sentiment')
plt.ylabel('Effectif')
plt.xticks(rotation=45)
plt.show()

# Histogramme pour les émotions
plt.figure(figsize=(10, 8))
emotions = ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']
data[emotions].sum().plot(kind='bar', color='lightcoral')
plt.title('Distribution des Émotions')
plt.xlabel('Émotion')
plt.ylabel('Effectif')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculer les moyennes des scores d'émotion
emotion_means = data[['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']].mean()

# Couleurs pour les émotions
emotion_colors = ['#FF5733', '#FFC300', '#C70039', '#900C3F', '#581845', '#1E8449']

# Tracer l'histogramme des moyennes des scores d'émotion avec des couleurs personnalisées
plt.figure(figsize=(10, 8))
emotion_means.plot(kind='bar', color=emotion_colors, edgecolor='black')
plt.title('Moyennes des Scores d\'Émotion')
plt.xlabel('Émotion')
plt.ylabel('Moyenne du Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Couleurs pour les sentiments
sentiment_colors = {'positive score': 'green', 'neutral score': 'blue', 'negative score': 'red'}

# Tracer l'histogramme des moyennes des scores de sentiment avec des couleurs personnalisées
plt.figure(figsize=(8, 6))
sentiment_means.plot(kind='bar', color=sentiment_means.index.map(sentiment_colors), edgecolor='black')
plt.title('Moyennes des Scores de Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Moyenne du Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Couleurs pour les émotions (pastel)
emotion_colors_pastel = ['#8E8E93', '#FFD700', '#FFB6C1', '#FFA07A', '#AFEEEE', '#98FB98']

# Tracer l'histogramme des moyennes des scores d'émotion avec des couleurs pastel
plt.figure(figsize=(10, 8))
emotion_means.plot(kind='bar', color=emotion_colors_pastel, edgecolor='black')
plt.title('Moyennes des Scores d\'Émotion')
plt.xlabel('Émotion')
plt.ylabel('Moyenne du Score')
plt.xticks(rotation=45)
plt.show()

# Couleurs pour les sentiments (pastel)
sentiment_colors_pastel = {'positive score': '#77DD77', 'neutral score': '#ADD8E6', 'negative score': '#FFB6C1'}

# Tracer l'histogramme des moyennes des scores de sentiment avec des couleurs pastel
plt.figure(figsize=(8, 6))
sentiment_means.plot(kind='bar', color=sentiment_means.index.map(sentiment_colors_pastel), edgecolor='black')
plt.title('Moyennes des Scores de Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Moyenne du Score')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Calculer les moyennes des scores d'émotion
emotion_means = data[['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']].mean().reset_index()
emotion_means.columns = ['Emotion', 'Proportion']

# Créer une liste de couleurs correspondant à chaque émotion
emotion_colors = {
    'sadness': '#8B0000',   # Rouge foncé
    'joy': '#228B22',       # Vert foncé
    'love': '#228B22',      # Vert foncé
    'anger': '#8B0000',     # Rouge foncé
    'fear': '#8B0000',      # Rouge foncé
    'surprise': '#228B22'   # Vert foncé
    # Ajoutez d'autres couleurs si nécessaire
}


# Remplacer les couleurs des émotions dans le DataFrame
emotion_means['Color'] = emotion_means['Emotion'].map(emotion_colors)

# Créer le graphique à barres avec les couleurs spécifiées
fig = px.bar(emotion_means, x='Emotion', y='Proportion',
             color='Color',  # Utiliser les couleurs spécifiées
             color_discrete_map=emotion_colors)  # Mappage des couleurs discrètes

# Modifier la mise en page pour un thème clair
fig.update_layout(
    title='Proportion of Each Emotion',
    plot_bgcolor='white',  # Arrière-plan du graphique en blanc
    paper_bgcolor='white', # Arrière-plan de la légende en blanc
    font_color='black',  # Texte en noir
    showlegend=False,  # Masquer la légende
    xaxis=dict(
        title='Emotion',
        titlefont_size=16,
        color='black'  # Texte de l'axe x en noir
    ),
    yaxis=dict(
        title='Proportion',
        titlefont_size=16,
        color='black'  # Texte de l'axe y en noir
    )
)

# Supprimer la légende de l'échelle de couleurs
fig.update_layout(coloraxis_showscale=False)

# Afficher le graphique
fig.show()
```

```python
import plotly.express as px

# Create a figure for the average emotion scores
fig_emotion = px.bar(emotion_means, x='Emotion', y='Proportion', color='Emotion',
                     color_discrete_sequence=emotion_colors_pastel,
                     title='Average emotions scores',
                     labels={'Proportion': 'Average score', 'Emotion': 'Emotion'})

# Modify margins and bar gap
fig_emotion.update_layout(bargap=0.2, margin=dict(l=20, r=20, t=50, b=20))

# Show the figure
fig_emotion.show()

# Create a figure for the average sentiment scores
fig_sentiment = px.bar(sentiment_means, x=sentiment_means.index, y=sentiment_means,
                       color=sentiment_means.index,
                       color_discrete_map=sentiment_colors_pastel,
                       title='Average sentiments scores',
                       labels={'value': 'Average score', 'index': 'Sentiment'})

# Modify margins and bar gap
fig_sentiment.update_layout(bargap=0.2, margin=dict(l=20, r=20, t=50, b=20))

# Show the figure
fig_sentiment.show()
```
