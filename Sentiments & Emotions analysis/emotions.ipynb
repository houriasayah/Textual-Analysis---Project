```python
import pandas as pd
# Chargement des données depuis le fichier CSV
data = pd.read_csv("test_df_en.csv", sep=";")
data = data.drop(data.columns[0], axis=1)
data
```

```python
data = data.iloc[9000:]
data
```

```python
import re
from transformers import pipeline

# Assurez-vous que la colonne 'lyrics' est de type str
data['lyrics'] = data['lyrics'].astype(str)

# Initialisation du pipeline pour la classification de texte avec le modèle spécifié
classifier = pipeline("text-classification", model='bhadresh-savani/distilbert-base-uncased-emotion', return_all_scores=True)

# Fonction pour découper le texte en morceaux de taille maximale sans couper les mots
def chunk_text(text, chunk_size=512):
    chunks = []
    current_chunk = ""
    words = re.findall(r'\b\w+\b', text)  # Trouver tous les mots dans le texte
    for word in words:
        if len(current_chunk) + len(word) < chunk_size:
            current_chunk += word + " "
        else:
            chunks.append(current_chunk.strip())
            current_chunk = word + " "
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

# Fonction pour prédire les émotions à partir des paroles
def predict_emotion(lyrics):
    chunks = chunk_text(lyrics)
    predictions = []
    for chunk in chunks:
        prediction = classifier(chunk)
        predictions.append(prediction)
    return predictions

# Fonction pour extraire les scores d'émotion de chaque prédiction
def extract_emotion_scores(predictions):
    emotion_scores = {}
    num_chunks = len(predictions)
    for chunk_predictions in predictions:
        for chunk in chunk_predictions:
            for emotion_data in chunk:
                label = emotion_data['label']
                score = emotion_data['score']
                emotion_scores[label] = emotion_scores.get(label, []) + [score]
    return emotion_scores

# Appliquer la fonction predict_emotion sur un échantillon aléatoire de 1000 lignes
data['emotion_prediction'] = data['lyrics'].apply(predict_emotion)

# Extraire les scores d'émotion de chaque prédiction
data['emotion_scores'] = data['emotion_prediction'].apply(extract_emotion_scores)

# Calculer les moyennes des scores d'émotion pour chaque chanson
for label in data['emotion_scores'].iloc[0].keys():
    data[label] = data['emotion_scores'].apply(lambda x: sum(x[label]) / len(x[label]))
```

```python
# Exporter le DataFrame vers un fichier CSV
data.to_csv("emotion_predictions11.csv", index=False)
```
